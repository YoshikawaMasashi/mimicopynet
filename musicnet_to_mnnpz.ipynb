```python
# python3では普通に開くとこの後で何らかのDecodeErrorになるが、下のようにひらけば大丈夫！(thanks to @marshi)
train_data = np.load(open('./musicnet.npz','rb'),encoding='latin1')
```

```python
import os
```

```python
os.makedirs('musicnet_data', exist_ok=True)
```

```python
for id in train_data.keys(): # id は文字列（数字だけど）
    print('processing: id =',id)
    os.makedirs('musicnet_data/'+str(id),exist_ok=True)
    x, y = train_data[id] # x: 波形 (ndarray<float: -1~1> (num_of_sample,))、y: 楽譜データ (intervaltree)
    in_npy = x
    intvl = 512
    out_sample = np.array(range(0,len(x),intvl))
    out_npy = np.zeros((128, len(out_sample)))
    # outを集計
    for i,s in enumerate(out_sample):
        nns = [n[2][1] for n in y[s]]
        for nn in nns:
            out_npy[nn,i] = 1.
    print(in_npy.shape, out_npy.shape, out_sample.shape)
    np.savez('musicnet_data/'+str(id)+'/data.npz', wave=in_npy, score=out_npy, score_sample=out_sample)
    # 注意点
    # in_npyとout_npyみたいに分けることはしなくなった
    # waveはμ-law適用していない（あとでFTする可能性とかも考えて）
```
